## Data Cleaning
1. Delete a column

```python
data = data.drop(columns=['Supportive Environment Score'])
```

```python
data.head(10)
```

2. delete a row
### delete the row with 'NaN' in the column 'DBN '

```python
data = data.dropna(subset=['DBN'])
```

```python
data.describe()
```

#### We dont replace missing data with 0, why?

## find the outliers
1. draw boxplot

```python
data.boxplot(grid=False, rot=45, fontsize=15)
```

### This graph doesnt make sense because the data are in different scale, and they should not be in the same graph.

Let's look at the first five columns only

```python
new_data1 = data[data.columns[1:5]]
new_data1.boxplot(grid=False, rot=45, fontsize=15)
```

```python
new_data2 = data[data.columns[6:10]]
new_data2.boxplot(grid=False, rot=45, fontsize=15)
```
